```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)
plt.show()
```

```python
# Creating a scatter plot
x = [5, 10, 15, 20, 25]
y = [10, 15, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# Creating a scatter plot and a line plot
x = [5, 10, 15, 20, 25]
y = [10, 15, 20, 25, 30]
plt.plot(x, y, color='green', label='Line Plot')
plt.scatter(x, y, color='red', marker='o', label='Scatter Plot')
plt.title("Combined Line and Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Line plot in the first subplot
ax1.plot(x, y)
ax1.set_title("Line Plot")

# Bar chart in the second subplot
ax2.bar(categories, values)
ax2.set_title("Bar Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the plots
x = [5, 10, 15, 20, 25]
y = [10, 15, 20, 25, 30]
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Create a 1x2 grid of subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: Line and Scatter Plot
ax1.plot(x, y, color='blue', label='Line Plot')
ax1.scatter(x, y, color='red', marker='o', label='Scatter Plot')
ax1.set_title("Line and Scatter Plot")
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax1.legend()

# Second subplot: Bar Chart
ax2.bar(categories, values, color='green')
ax2.set_title("Bar Chart")
ax2.set_xlabel("Categories")
ax2.set_ylabel("Values")

# Display the plot
plt.tight_layout()  # Adjusts spacing to prevent overlap
plt.show()
```

```python
# Real-World Scenario (Grouped Bar Chart for Quarterly Profits)
import numpy as np

quarters = ['Q1', 'Q2', 'Q3', 'Q4']
profits_2022 = [10000, 12000, 11000, 13000]
profits_2023 = [15000, 14000, 16000, 17000]

x = np.arange(len(quarters))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x - width/2, profits_2022, width, label='2022', color='yellow')
ax.bar(x + width/2, profits_2023, width, label='2023', color='green')

ax.set_xticks(x)
ax.set_xticklabels(quarters)
ax.set_xlabel("Quarters")
ax.set_ylabel("Profits")
ax.set_title("Quarterly Profits Comparison")
ax.legend()

plt.show()
```
